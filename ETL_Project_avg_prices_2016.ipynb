```python
#Rename symbol column to Ticker symbol
company_prices.rename(columns={'symbol': 'Ticker symbol'}, inplace=True)
company_prices.head()
```

```python
#filter 2016 only
company_prices_2 = company_prices.loc[(company_prices['date'] >= '2016-01-01 00:00:00') & (company_prices['date'] <= '2016-12-31 00:00:00')]
company_prices_2.head()
#companies.loc['2016-01-01':'2016-12-31']
```

```python
#Merge company df and company_prices_2 df in order to do groupby Security
company_merge_company_prices_2 = pd.merge(company, company_prices_2, how="left", on=["Ticker symbol"])
company_merge_company_prices_2
```

```python
companies2 = company_merge_company_prices_2[['Ticker symbol','Security','close']].copy()
companies2
```

```python
avg_prices_2016 = companies2.groupby(['Ticker symbol']).mean()
avg_prices_2016.head()
```
